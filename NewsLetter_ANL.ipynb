## 📌 문제 정의

### 좋은 글이란 무엇인가?

뉴스레터 독자들의 반응과 텍스트 특성을 분석하여 **'좋은 글'의 기준을 데이터 기반으로 정의**하고자 한다.

---

## 📖 분석 배경

메일리(Maily) 뉴스레터 플랫폼을 운영하는 지인의 뉴스레터 데이터를 활용하여,  
독자들에게 높은 반응을 이끌어내는 글의 특징을 탐구하는 **텍스트 데이터 분석 프로젝트**를 진행하였다.

---

최종적으로는 새로운 뉴스레터 데이터가 업데이트될 때마다 해당 글을 평가하여,

- ✅ 충족(좋은 글)
- ⚠️ 주의(좋은 글과 나쁜글의 평균)

로 분류할 수 있는 **뉴스레터 품질 평가 시스템 구축**을 목표로 한다.

# 1. 📊 메일리 뉴스레터 평가지표 정의

## 개요

뉴스레터의 성과를 객관적으로 평가하기 위해 메일리(Maily) 플랫폼에서 제공하는 데이터를 분석하였다.

평가 지표는 크게 다음 세 영역으로 구분된다.

| 구분     | 목적                |
| ------ | ----------------- |
| 이메일 통계 | 구독자의 직접적인 반응 측정   |
| 웹 통계   | 외부 유입 및 콘텐츠 확산 측정 |
| 발송 내역  | 개별 구독자 행동 분석      |

---

# 1️⃣ 이메일 통계 (Email Analytics)

> 구독자가 뉴스레터 이메일과 어떻게 상호작용하는지 측정하는 핵심 성과 지표

### 📧 발송 (Delivery)

뉴스레터가 구독자의 이메일 서버로 성공적으로 전달된 횟수이다.

예시

```
83 / 85
```

* 83 : 실제 발송 성공 건수
* 85 : 해당 카테고리 구독자 수
* 차이(2명) : 이메일 주소 오류, 서버 문제 등으로 인한 발송 실패(Bounce)

---

### 📬 오픈 (Open)

뉴스레터를 열람한 고유(Unique) 구독자 수이다.

* 동일 사용자가 여러 번 열람해도 1회로 계산
* 뉴스레터의 기본적인 관심도를 측정하는 대표 지표

---

### 🖱 클릭 (Click)

뉴스레터 본문 내 링크를 1회 이상 클릭한 고유 구독자 수이다.

* 단순 열람을 넘어 실제 행동(Action)으로 이어진 비율 확인 가능
* 콘텐츠의 설득력과 CTA(Call To Action) 효과 측정 가능

---

### ❌ 구독 취소 (Unsubscribe)

뉴스레터 수신 후 구독을 해지한 인원 수이다.

* 콘텐츠 만족도를 간접적으로 평가하는 지표
* 높은 수치는 콘텐츠 피로도 또는 타겟 부적합 가능성을 의미

---

### 📈 이메일 오픈 추이

시간 흐름에 따른 전체 열람 횟수 변화를 나타낸다.

* 재오픈(중복 열람) 포함
* 콘텐츠의 반복 소비 가치를 평가 가능

---

### ⏰ 시간대별 오픈 추이

발송 이후 어느 시간대에 가장 많이 열람되는지 보여주는 지표이다.

활용 목적

* 최적 발송 시간 탐색
* 구독자 행동 패턴 분석

---

### 🔗 링크 클릭 지표

본문 내 개별 링크의 클릭 수를 측정한다.

분석 가능 항목

* 가장 높은 반응을 얻은 CTA
* 독자가 관심을 보인 콘텐츠 영역
* 전환 성과가 높은 링크 유형

※ 유저 저니(User Journey) 데이터와 연계하여 검증하였다.



[이메일 통계] 링크 클릭 지표 클릭 초기 값 3회→ 유저 저니 경험 클릭 2회 총 5회 변경 확인됨

---

# 2️⃣ 웹 통계 (Web Analytics)

> 이메일 외부에서 발생하는 콘텐츠 유입과 확산 성과 측정

### 🌐 웹 조회수 (Page Views)

뉴스레터 웹 페이지(아카이브)의 총 조회 수이다.

특징

* 중복 방문 포함
* 새로고침 및 재방문도 모두 집계

따라서 이메일 오픈 수보다 콘텐츠 소비량을 더 폭넓게 측정할 수 있다.

---

### 🚪 유입 경로 (Referral)

방문자가 어떤 경로를 통해 뉴스레터에 접근했는지 보여주는 데이터이다.

예시

* Google 검색
* 카카오톡 공유
* Facebook
* 블로그 링크
* 직접 접속

이를 통해 효과적인 홍보 채널을 파악할 수 있다.

---

### 📅 웹 오픈 추이

날짜별 웹 조회 수 변화를 보여주는 지표이다.

활용 목적

* 콘텐츠의 장기적 생명력 평가
* 검색 유입 효과 확인
* 장기 조회(롱테일 트래픽) 분석

---

### 🎯 주요 리퍼럴 지표

5회 이상 유입이 발생한 주요 유입 경로만 선별하여 제공한다.

특징

* 우연적 유입 제거
* 의미 있는 유입 채널 식별
* 마케팅 성과 분석에 활용

---

# 3️⃣ 발송 내역 (Delivery Log)

> 개별 구독자 단위의 행동 로그(Log) 데이터

### 👤 구독자

발송 시점 기준 전체 수신 대상 이메일 정보이다.

---

### 🏷 구독자 종류

구독자 유형을 구분하는 항목이다.

예시

* 일반 구독자
* 그룹 전용 구독자
* 유료 구독자

---

### 📌 구독 현황

구독자의 현재 상태를 나타낸다.

* 구독 중
* 구독 취소

---

### 📨 발송 상태

개별 이메일의 전달 결과를 확인할 수 있다.

예시

* 발송 성공

* 오픈 완료

---

### 🕒 발송 시간

---

### 👀 오픈 기록

구독자가 이메일을 열람한 모든 시각이 기록된다.

분석 가능 항목

* 최초 열람 시점
* 재열람 여부
* 콘텐츠 관심도
* 충성 구독자 식별

---

### 🖱 클릭 기록

구독자가 본문 링크를 클릭한 모든 시각이 기록된다.

분석 가능 항목

* 행동 전환 시점

# 2. 🎯 문제 정의 기반 카테고리별 메일리 통계 지표 변수 선정

## 개요

본 프로젝트의 핵심 연구 질문은 다음과 같다.

> **"좋은 글이란 무엇인가?"**

뉴스레터 데이터를 검토한 결과, 에디터의 콘텐츠는 크게 **The Trail**과 **Pilgrimage** 카테고리로 구분되어 있었다.

두 카테고리는 다루는 소재와 독자가 기대하는 경험에 차이가 있을 수 있다고 판단하였다.

예를 들어 The Trail은 여행 경험과 개인적 사유를 중심으로 한 에세이 성격이 강하며, Pilgrimage는 인물과 브랜드 사례를 통해 인사이트를 전달하는 콘텐츠의 비중이 높았다.

이에 따라 본 프로젝트에서는 모든 콘텐츠를 동일한 기준으로 평가하기보다, 각 카테고리의 특성을 고려하여 '좋은 글'의 기준을 설정하고 그에 적합한 성과 지표를 선정하였다.

선정된 지표는 절대적인 기준이라기보다, 각 카테고리의 특성을 설명하는 데 상대적으로 적합하다고 판단한 변수들이다.

---

## 2-1. ✈️ The Trail (여행 · 에세이)

### 좋은 글의 정의

> 독자의 감정을 움직이고, 읽은 이후에도 여운이 남아 다시 찾아보거나 타인에게 공유하고 싶게 만드는 글

The Trail은 여행 경험과 개인적 사유를 중심으로 한 에세이 성격의 콘텐츠이다.

따라서 단순 클릭 수보다 독자가 얼마나 깊게 몰입하고 공감했는지가 더 중요한 평가 기준이 된다.

---

### 📌 핵심 변수 ① 다회오픈율 (Multi-Open Rate)

#### 계산식

```text
다회오픈율 = 2회 이상 오픈 기록이 있는 구독자 수 ÷ 전체 오픈 수
```

> 변수 정의: 1장에서 정의한 이메일 통계의 오픈(Open) 과 발송 내역의 오픈 기록(Open Log) 을 결합하여 생성한 지표이다. 앞으로 나올 지표 또한 1장에서 정의한 이메일 통계 지표를 활용하였다.

#### 선정 이유

에세이를 여러 번 다시 읽는 행동은 정보 탐색 목적보다는 문장 자체에 대한 감상과 공감에서 발생한다.

즉, 독자가 반복적으로 콘텐츠를 열람했다는 것은 글이 감정적 울림을 제공했다는 의미로 해석할 수 있다.

#### 기대 효과

다회오픈율이 높은 글을 분석함으로써 독자가 오래 기억하는 문체적 특징을 도출할 수 있다.

---

### 📌 핵심 변수 ② 바이럴 비율 (Viral Ratio)

#### 계산식

```text
바이럴 비율 = 웹 조회수 ÷ 이메일 조회수
```

#### 선정 이유

독자가 다른 사람에게 콘텐츠를 공유하는 행동은

> "나만 보기 아까운 글"

이라고 판단했다는 의미로 해석할 수 있다.

따라서 바이럴 비율은 콘텐츠의 공감 확산력과 서사적 힘을 측정하는 지표로 활용할 수 있다.

#### 기대 효과

바이럴 비율이 높은 글을 분석하여 독자들이 타인과 공유하고 싶어하는 서사 구조를 파악할 수 있다.

---

## 2-2. 👤 Pilgrimage (인물 · 브랜드)

### 좋은 글의 정의

> 독자에게 새로운 관점과 인사이트를 제공하고 행동을 유도하는 글

Pilgrimage는 특정 인물과 브랜드의 성공 과정, 전략, 철학 등을 다루는 콘텐츠이다.

따라서 감성적 여운보다는 독자가 얼마나 설득되었고 실제 행동으로 이어졌는지가 중요한 평가 기준이 된다.

---

### 📌 핵심 변수 ① CTOR (Click-To-Open Rate)

#### 계산식

```text
CTOR = 이메일 클릭 수 ÷ 이메일 오픈 수
```

#### 선정 이유

독자가 이메일을 열었다고 해서 반드시 콘텐츠를 긍정적으로 평가했다고 볼 수는 없다.

반면 링크 클릭은 독자가 콘텐츠에 충분한 가치를 느껴 추가 행동을 수행했다는 의미를 가진다.

따라서 CTOR은 콘텐츠의 설득력과 인사이트 전달력을 평가할 수 있는 핵심 지표이다.

#### 기대 효과

CTOR이 높은 글을 분석하여 독자의 행동을 유도하는 스토리텔링 구조를 파악할 수 있다.

---

### 📌 핵심 변수 ② 오픈율 (Open Rate)

#### 계산식

```text
오픈율 = 이메일 오픈 수 ÷ 이메일 발송 수
```

#### 선정 이유

인물 및 브랜드 콘텐츠는 제목과 소재 선택이 성과에 큰 영향을 미친다.

내용이 우수하더라도 독자의 관심을 끌지 못하면 읽히지 않기 때문이다.

따라서 오픈율은 콘텐츠 자체보다도 독자의 호기심을 자극하는 프레이밍 능력을 평가하는 지표로 활용할 수 있다.

#### 기대 효과

오픈율이 높은 글을 분석하여 독자의 관심을 끌어내는 소재 선정 및 제목 작성 전략을 도출할 수 있다.

# 4. 📈 CSV 데이터 기반 카테고리별 변수 상관관계 분석


## 개요

3장에서 구축한 CSV 데이터셋을 활용하여 카테고리별 핵심 성과 지표 간의 관계를 점도표(Scatter Plot)로 시각화하였다.

이를 통해 각 변수들이 실제로 함께 움직이는지 확인하고, 최종 텍스트 분석 및 예측 모델에 활용할 변수를 선정하고자 하였다.

---


````python
# 데이터 처리를 위한 pandas와 데이터 시각화를 위한 matplotlib 라이브러리를 불러옵니다.
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# 데이터 불러오기
# =========================
# 한글 깨짐을 방지하기 위해 'cp949' 인코딩을 사용하여 CSV 파일을 읽고 데이터프레임(df)으로 저장합니다.
df = pd.read_csv('뉴스레터_통계.csv', encoding='cp949')

# =========================
# 파생 변수 생성
# =========================
# 이메일 오픈(조회) 대비 클릭 비율인 CTOR(Click-To-Open Rate)을 백분율로 계산하여 새 열로 추가합니다.
df['CTOR'] = df['이메일 클릭'] / df['이메일 조회'] * 100
# 이메일 조회 대비 웹 조회 비율을 계산하여 콘텐츠 외부 확산도를 보여주는 Viral Ratio 열을 추가합니다.
df['Viral Ratio'] = df['웹 조회'] / df['이메일 조회'] * 100

# =========================
# 색상 및 범례
# =========================

# (바이럴 vs 다회오픈율)
# 콘텐츠의 파급력과 여운을 평가합니다.
# Masterpiece (확산O, 여운O): 널리 퍼지고 깊이 읽힌 명작. 브랜드의 자산이 되는 글.
# Cult Classic (확산X, 여운O): 대중적이지 않지만 팬덤이 열광하는 글. 충성 독자를 만듦.
# Snack (확산O, 여운X): 가볍게 소비되고 빠르게 퍼지는 글. 휘발성이 강함.
# Soso (확산X, 여운X): 조용히 소비되고 잊혀지는 글.

# 'The Trail' 카테고리의 4사분면 분류에 매칭될 색상을 딕셔너리로 정의합니다.
trail_colors = {
    'Masterpiece': 'forestgreen',
    'Cult Classic': 'royalblue',
    'Snack': 'orange',
    'Soso': 'lightgray'
}

# 범례(Legend)에 표시될 라벨 이름을 딕셔너리로 정의합니다.
trail_labels = {
    'Masterpiece': 'Masterpiece',
    'Cult Classic': 'Cult Classic',
    'Snack': 'Snack',
    'Soso': 'Soso'
}

# (오픈율 vs CTOR)
# 제목과 본문의 매력도를 평가합니다.
# Star (제목O, 내용O): 가장 이상적인 글. 제목 어그로도 성공했고, 내용도 독자를 만족시킴.
# Hidden Gem (제목X, 내용O): 내용은 훌륭하나 제목이 아쉬워 묻힌 글. 제목 리라이팅 1순위.
# Clickbait (제목O, 내용X): 제목은 잘 뽑았으나 내용이 기대에 못 미친 '낚시성' 글. CTOR 개선 필요.
# Underperformer (제목X, 내용X): 총체적 난국. 기획 단계부터 재검토 필요.

# 'Pilgrimage' 카테고리의 4사분면 분류에 매칭될 색상을 정의합니다.
pil_colors = {
    'Star': 'green',
    'Hidden Gem': 'royalblue',
    'Clickbait': 'orange',
    'Underperformer': 'lightgray'
}

# 범례에 표시될 라벨 이름을 정의합니다.
pil_labels = {
    'Star': 'Star',
    'Hidden Gem': 'Hidden Gem',
    'Clickbait': 'Clickbait',
    'Underperformer': 'Underperformer'
}

# =========================
# Figure 생성
# =========================
# 그래프의 전체적인 디자인 테마를 'ggplot' 스타일로 설정합니다.
plt.style.use('ggplot')

# 가로 15, 세로 6 크기의 1행 2열(그래프 2개)짜리 전체 도화지(fig)와 각각의 그래프 영역(axes)을 생성합니다.
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

###################################################
# LEFT : The Trail
###################################################

# 왼쪽 첫 번째 그래프 영역을 ax로 지정합니다.
ax = axes[0]

# 원본 데이터에서 카테고리가 'The Trail'인 행들만 필터링하여 독립적인 복사본(trail)을 만듭니다.
trail = df[df['카테고리'] == 'The Trail'].copy()

# x축과 y축에 사용할 컬럼명을 변수로 지정합니다.
x_col = 'Viral Ratio'
y_col = '다회오픈율'

# 4사분면을 나누는 기준점이 될 x축과 y축 데이터의 중앙값(median)을 각각 구합니다.
x_med = trail[x_col].median()
y_med = trail[y_col].median()

# 중앙값을 기준으로 데이터를 4가지 성과 그룹 중 하나로 분류하는 함수입니다.
def classify_trail(row):
    if row[x_col] >= x_med and row[y_col] >= y_med:
        return 'Masterpiece'
    elif row[x_col] < x_med and row[y_col] >= y_med:
        return 'Cult Classic'
    elif row[x_col] >= x_med and row[y_col] < y_med:
        return 'Snack'
    else:
        return 'Soso'

# 정의한 분류 함수를 trail 데이터의 각 행에 적용(axis=1)하여 'Category'라는 새로운 열에 결과를 저장합니다.
trail['Category'] = trail.apply(classify_trail, axis=1)

# 미리 정의한 색상 딕셔너리(trail_colors)를 순회하며 그룹별로 산점도(scatter)를 그립니다.
for group, color in trail_colors.items():

    # 현재 반복 중인 그룹에 해당하는 데이터만 추출합니다.
    data = trail[trail['Category'] == group]

    # 추출한 그룹 데이터를 그래프에 점으로 표시합니다. (크기, 색상, 테두리 등 설정)
    ax.scatter(
        data[x_col],
        data[y_col],
        s=140,
        color=color,
        edgecolor='black',
        linewidth=1.2,
        alpha=0.85,
        label=trail_labels[group]
    )

# T1, T2... 등 각 점의 식별자를 표시하기 위해 전체 행을 반복합니다.
for _, row in trail.iterrows():

    # 각 데이터 점의 x좌표에서 우측으로 살짝(0.3) 이동한 위치에 인덱스(Index) 텍스트를 출력합니다.
    ax.text(
        row[x_col] + 0.3,
        row[y_col],
        row['Index'],
        fontsize=9
    )

# 기준점이 되는 중앙값 위치에 투명도가 있는 점선(dashed line)을 그려 십자선 형태로 사분면을 나눕니다.
ax.axvline(x_med, linestyle='--', color='gray', alpha=0.5)
ax.axhline(y_med, linestyle='--', color='gray', alpha=0.5)

# 왼쪽 그래프의 제목, x축 이름(중앙값 포함), y축 이름(중앙값 포함)을 설정합니다.
ax.set_title('The Trail: Viral Ratio vs Multi_Open Rate')
ax.set_xlabel(f'Viral Ratio (Median {x_med:.1f}%)')
ax.set_ylabel(f'Multi_Open Rate (Median {y_med:.1f}%)')

# 그룹별 색상 정보가 담긴 범례를 화면에 표시합니다.
ax.legend()

###################################################
# RIGHT : Pilgrimage
###################################################

# 오른쪽 두 번째 그래프 영역을 ax로 지정합니다.
ax = axes[1]

# 원본 데이터에서 카테고리가 'Pilgrimage'인 행들만 필터링하여 복사본(pil)을 만듭니다.
pil = df[df['카테고리'] == 'Pilgrimage'].copy()

# 오른쪽 그래프에 맞게 x축('클릭률')과 y축('CTOR') 변수를 설정합니다.
x_col = '클릭률'
y_col = 'CTOR'

# 기준선이 될 x축과 y축 데이터의 중앙값을 구합니다.
x_med = pil[x_col].median()
y_med = pil[y_col].median()

# Pilgrimage 전용 중앙값 기준으로 4가지 카테고리로 분류하는 함수를 선언합니다.
def classify_pil(row):
    if row[x_col] >= x_med and row[y_col] >= y_med:
        return 'Star'
    elif row[x_col] < x_med and row[y_col] >= y_med:
        return 'Hidden Gem'
    elif row[x_col] >= x_med and row[y_col] < y_med:
        return 'Clickbait'
    else:
        return 'Underperformer'

# 분류 함수를 pil 데이터 각 행에 적용하여 'Category' 열을 채웁니다.
pil['Category'] = pil.apply(classify_pil, axis=1)

# 설정된 색상 딕셔너리(pil_colors)에 따라 그룹별 데이터를 순회하며 산점도를 그립니다.
for group, color in pil_colors.items():

    # 현재 그룹과 일치하는 데이터만 필터링합니다.
    data = pil[pil['Category'] == group]

    # 산점도를 시각화하여 찍어줍니다.
    ax.scatter(
        data[x_col],
        data[y_col],
        s=140,
        color=color,
        edgecolor='black',
        linewidth=1.2,
        alpha=0.85,
        label=pil_labels[group]
    )

# P1, P2... 등 포인트 라벨링을 위해 반복문을 실행합니다.
for _, row in pil.iterrows():

    # 점의 겹침을 방지하기 위해 x좌표를 살짝 띄워 인덱스 텍스트를 출력합니다.
    ax.text(
        row[x_col] + 0.3,
        row[y_col],
        row['Index'],
        fontsize=9
    )

# 오른쪽 그래프에도 기준이 되는 중앙값 십자선(점선)을 추가합니다.
ax.axvline(x_med, linestyle='--', color='gray', alpha=0.5)
ax.axhline(y_med, linestyle='--', color='gray', alpha=0.5)

# 오른쪽 그래프의 제목과 축 라벨을 설정합니다.
ax.set_title('Pilgrimage: Open Rate vs CTOR')
ax.set_xlabel(f'Open Rate (Median {x_med:.1f}%)')
ax.set_ylabel(f'CTOR (Median {y_med:.1f}%)')

# 우측 그래프 범례를 표시합니다.
ax.legend()

# 서브플롯 간의 간격과 텍스트가 서로 겹치지 않도록 전체 레이아웃을 자동으로 다듬어줍니다.
plt.tight_layout()
# 최종적으로 완성된 두 그래프를 화면에 출력합니다.
plt.show()
````

## 4-2. ✈️ The Trail 카테고리 분석

### 분석 변수

- 바이럴 비율 (Viral Ratio)
- 다회오픈율 (Multi-Open Rate)

### 분석 결과

대부분의 콘텐츠는 바이럴 비율이 증가할수록 다회오픈율도 함께 증가하는 경향을 보였다.

이는 공유가 활발한 콘텐츠가 반복 열람에서도 일정 수준의 성과를 보일 수 있음을 시사한다.

---

### 데이터 분포 특징

#### ① 주요 군집의 우상향 경향

대부분의 콘텐츠는 특정 구간 내에서 비교적 안정적인 우상향 패턴을 형성하였다.

이는 공유와 재열람이 일정 부분 함께 발생하는 경향이 있음을 보여준다.

#### ② 이질적인 이상치(Outlier) 존재

일부 콘텐츠는 다회오픈율이 높지 않음에도 매우 높은 바이럴 비율을 기록하였다.

반대로 다회오픈율이 높은 콘텐츠가 반드시 높은 바이럴 비율을 보이지는 않았다.

이러한 결과는 콘텐츠 소비가

- 공유 중심 소비(Viral)
- 반복 열람 중심 소비(Deep Reading)

라는 서로 다른 형태로 나타날 가능성을 시사한다.

---

## ⚠️ 결과 해석 시 유의사항

The Trail 카테고리의 바이럴 비율은 전체적으로 매우 높은 수치를 기록하였다.

해당 수치가 실제 이메일 구독자로부터 공유를 하여 바이럴을 타고 콘텐츠 확산 효과로 완전히 반영되었다고 단정하기는 어렵다는 판단을 내렸다.

바이럴 비율은 참고 지표로 활용하되, 실제 독자 행동을 보다 정밀하게 이해하기 위해서는 추가적으로 유저 저니 분석이 다시 필요하다고 판단하였다.

# 5. 🔍 바이럴 지표 유저 저니 심층 분석 및 최종 모델 활용 변수 선정

## 개요

4장에서 The Trail 카테고리의 핵심 변수 후보로 선정했던 **바이럴 비율(Viral Ratio)** 과 **다회오픈율(Multi-Open Rate)** 에 대해 추가 검증을 수행하였다.

특히 바이럴 비율의 경우 실제로 "좋은 글"의 확산력을 측정하는 지표로 활용 가능한지 확인하기 위해 유저 저니(User Journey)를 기반으로 지표의 타당성을 재검토하였다.

---

## 📌 바이럴 지표 검증 배경

초기 분석에서는 다음과 같은 계산식을 활용하였다.

```text
Viral Ratio = 웹 조회수 ÷ 이메일 조회수
```

해당 지표는 다음과 같은 가정을 기반으로 설계되었다.

> 이메일을 읽은 구독자가 콘텐츠를 공유하고, 이를 통해 외부 방문자가 유입될수록 좋은 글일 가능성이 높다.

그러나 실제로 이 지표가 콘텐츠의 공유와 확산을 측정하는지 확인하기 위해 추가 검증을 진행하였다.

## 5-1. 🚨 기존 바이럴 지표의 한계


### 문제점 ① 공유 과정 추적 한계

기존 바이럴 비율은 이메일 조회 이후 발생하는 행동 경로를 직접 확인할 수 없었다.

예시)

```text
이메일 조회
→ 링크 클릭
→ 웹 페이지 이동
→ URL 복사
→ 외부 공유
→ 신규 방문자 유입
```

메일리 통계에서는 위 과정 중 일부 단계만 확인 가능하며, 실제 공유가 발생했는지 여부를 확인하기 어려웠다.

---

### 문제점 ② 웹 조회수의 의미 왜곡 가능성

웹 조회수에는 다양한 유입 경로가 함께 포함된다.

예시

* 구글 검색
* 카카오톡 공유
* SNS 링크
* 직접 접속
* 기존 방문자의 재방문

따라서 웹 조회수 증가가 반드시 뉴스레터 구독자의 공유 때문이라고 해석하기 어렵다.

# 6. 📝 텍스트 데이터 분석에 사용되는 뉴스레터 콘텐츠 선정

## 개요

본 프로젝트의 목적은 뉴스레터 성과 지표와 텍스트 특성 간의 관계를 분석하여 "좋은 글"의 특징을 탐색하는 것이다.

이를 위해 카테고리별로 선정한 핵심 성과 지표를 기준으로 뉴스레터를 분류하고, 성과가 상대적으로 높은 콘텐츠(Good)와 낮은 콘텐츠(Bad)를 비교 분석하였다.

예시)

```text
P1 : 당신과 다른 트럼프의 어린시절 (1/4)
```

# 7. 🔬 Good·Bad 뉴스레터 비교 분석 프레임워크 설계

## 개요

6장에서 선정한 Good 콘텐츠와 Bad 콘텐츠를 대상으로 텍스트 데이터 분석을 수행하였다.

본 프로젝트는 총 16개의 뉴스레터를 분석 대상으로 하며, 콘텐츠별 길이 편차가 크고 표본 수가 제한적이다. 따라서 일반적인 평균 비교나 회귀분석 중심 접근보다 사례 연구(Case Study)와 심층 프로파일링(Deep Profiling) 관점의 분석 프레임워크를 설계하였다.

---

## 7-1. 🎯 분석 방향 설정

### 기존 통계 분석 접근의 한계

본 프로젝트는 다음과 같은 특성을 가진다.

- 소표본 데이터 (N=16)
- 콘텐츠별 길이 편차 존재
- 카테고리별 성격 차이 존재

따라서 단순 평균 비교나 빈도 중심 분석은 개별 콘텐츠의 특성을 충분히 설명하기 어렵다고 판단하였다.

---

### 분석 관점 전환

> **"평균(Average)이 아닌 밀도(Density)와 흐름(Flow)을 본다."**

모든 텍스트는 길이 정규화 과정을 거친 후 비교하며, 정량 지표와 정성적 해석을 함께 활용하는 분석 방식을 채택하였다.

---

## 7-2. 🧩 최종 분석 지표 선정

텍스트 분석 도구 검토 결과, 데이터 규모와 분석 목적을 고려하여 다음 지표를 최종 활용하였다.

| 분석 영역 | 활용 지표 |
|-----------|-----------|
| 구조적 완성도 | 담화 응집성 |
| 감정의 서사 | 평균 감성, 감성 표준편차, 마지막 긍정 마감 |
| 언어적 품질 | MSTTR |
| 문체 특성 | 활동성, 참여도 |

반면 TF-IDF, LDA 등 빈도 기반·토픽 기반 분석은 데이터 규모의 한계로 인해 활용하지 않았다.

## 7-5. 📊 텍스트 특성 지표 정의 및 수치화

뉴스레터의 문체적 특성을 분석하기 위해 **다양성, 활동성, 참여도** 지표를 추가적으로 산출하였다.

---

### 🔤 다양성 (MSTTR)

#### 정의

어휘 사용의 풍부함과 다양성을 측정하는 지표

#### 산출 로직

```text
전체 텍스트
→ 500 Token 단위 분할
→ Segment별 TTR 계산
→ 평균 산출(MSTTR)
```

```text
TTR = 고유 어휘 수 ÷ 전체 어절 수
```

#### 활용 목적

- 어휘 다양성 비교
- 반복 표현 사용 여부 확인

---

### ⚡ 활동성 (Activeness)

#### 정의

전체 형태소 중 동사와 형용사가 차지하는 비율

#### 산출 로직

```text
활동성
=
(동사 수 + 형용사 수)
÷ 전체 형태소 수
```

#### 활용 목적

- 서사적·묘사적 표현 수준 측정
- 정보 전달형 문체와 비교

---

### 🙋 참여도 (Reader Engagement)

#### 정의

독자를 지칭하거나 행동을 유도하는 표현의 비율

#### 측정 대상

- 대명사 (당신, 여러분, 우리 등)
- 청유 표현 (생각해봅시다, 살펴보겠습니다 등)
- 의문 표현 (왜일까?, 어떻게 가능했을까? 등)

#### 산출 로직

```text
참여도
=
(대명사 + 청유 표현 + 의문 표현 수)
÷ 전체 문장 수
```

#### 활용 목적

- 독자와의 상호작용 수준 측정
- 참여 유도형 문체 여부 확인

---

### 📌 지표 요약

| 지표 | 정의 | 의미 |
|------|------|------|
| MSTTR | 보정된 어휘 다양성 | 언어적 풍부함 |
| 활동성 | 동사·형용사 비율 | 서사성·생동감 |
| 참여도 | 독자 지향 표현 비율 | 독자 상호작용 |

````python
# 한국어 형태소 분석을 위한 라이브러리를 불러옵니다.
from kiwipiepy import Kiwi
# 파일 존재 여부 등 운영체제 기능 사용을 위해 불러옵니다.
import os
# 배열 연산 및 평균값(MSTTR) 계산을 위해 넘파이를 불러옵니다.
import numpy as np

def analyze_multiple_newsletters():
    # 형태소 분석기(Kiwi) 객체를 생성하여 준비합니다.
    kiwi = Kiwi()
    
    # 분석 대상 파일 리스트 (P시리즈와 T시리즈 모두 포함)
    file_names = [
        "P1.txt", "P5.txt", "P6.txt", "P9.txt", "P11.txt", "P13.txt", "P14.txt", "P15.txt",
        "T1.txt", "T2.txt", "T4.txt", "T8.txt", "T9.txt", "T16.txt", "T18.txt", "T20.txt"
    ]
    
    # 각 텍스트 파일의 분석 결과(딕셔너리 형태)를 차곡차곡 모아둘 빈 리스트를 생성합니다.
    results = []

    print(f"🚀 총 {len(file_names)}개의 뉴스레터 분석을 시작합니다.\n")

    # 지정된 파일 리스트를 하나씩 순회하며 분석을 시작합니다.
    for file_name in file_names:
        # 파일이 실제 경로에 존재하지 않으면 경고 메시지를 띄우고 다음 파일로 넘어갑니다.
        if not os.path.exists(file_name):
            print(f"⚠️ {file_name} 파일을 찾을 수 없어 건너뜁니다.")
            continue

        try:
            # 파일을 읽기 모드('r')와 utf-8 인코딩으로 열어 전체 텍스트를 읽어옵니다.
            with open(file_name, 'r', encoding='utf-8') as f:
                content = f.read()

            # [사전 작업] 불용어 처리
            # 분석에 방해되는 "첨부 이미지"라는 단어를 지우고, 양끝의 공백을 제거합니다.
            clean_text = content.replace("첨부 이미지", "").strip()
            
            # 내용이 비어있다면 분석할 것이 없으므로 다음 파일로 넘어갑니다.
            if not clean_text:
                continue

            # 형태소 분석
            # 정제된 텍스트를 형태소 단위로 쪼갭니다(토큰화).
            tokens = kiwi.tokenize(clean_text)
            total_tokens = len(tokens)

            # 1. 활동성 지표
            # 동사(VV)와 형용사(VA)를 활동성을 나타내는 품사로 정의합니다.
            dynamic_tags = ('VV', 'VA')
            # 전체 토큰 중 동사/형용사로 시작하는 태그를 가진 토큰의 개수를 셉니다.
            dynamic_count = len([t for t in tokens if t.tag.startswith(dynamic_tags)])
            # 전체 단어 중 활동성 단어가 차지하는 비중(%)을 계산합니다.
            activity_index = (dynamic_count / total_tokens) * 100 if total_tokens > 0 else 0

            # 2. 어휘 다양성 (MSTTR - 윈도우 500)
            # 텍스트 길이에 따른 편차를 줄이기 위해 500단어씩 묶어서(윈도우) 어휘 다양성을 측정합니다.
            window_size = 500
            if total_tokens < window_size:
                # 총 단어 수가 500개 미만이면 전체 텍스트에 대해 중복 없는 단어(Type) 비율을 구합니다.
                msttr_score = (len(set(t.form for t in tokens)) / total_tokens) * 100
            else:
                ttr_list = []
                # 전체 토큰을 500개 단위의 묶음(세그먼트)으로 나눕니다.
                for i in range(total_tokens // window_size):
                    seg = tokens[i*window_size : (i+1)*window_size]
                    # 각 묶음 안에서 중복 없는 단어의 비율을 구해 리스트에 추가합니다.
                    ttr_list.append(len(set(t.form for t in seg)) / window_size)
                # 모든 묶음의 다양성 비율 평균을 내어 최종 MSTTR 점수(%)를 산출합니다.
                msttr_score = np.mean(ttr_list) * 100

            # 3. 독자 참여 지표 (NP, EC, EF 조건)
            # 독자를 지칭하는 대명사(NP), 청유형 어미(EC), 의문형 종결어미(EF)를 지정합니다.
            target_np = {"우리", "너", "그대", "자네", "당신", "너희", "여러분"}
            target_ec = {"ㅂ시다", "ㄹ까요"}
            target_ef = {"ㄹ래요", "ㄹ까", "ㄹ까요", "나요", "ㄴ가", "ㅂ니까", "는가", "을까", "을까요", "ㄴ가요"}
            
            eng_count = 0
            # 모든 토큰을 확인하며 위 조건에 해당하는 단어가 몇 번 쓰였는지 카운트합니다.
            for t in tokens:
                if (t.tag == 'NP' and t.form in target_np) or \
                   (t.tag == 'EC' and t.form in target_ec) or \
                   (t.tag == 'EF' and t.form in target_ef):
                    eng_count += 1
            
            # 전체 단어 중 독자 참여를 유도하는 단어의 비중(%)을 계산합니다.
            engagement_index = (eng_count / total_tokens) * 100 if total_tokens > 0 else 0

            # 결과 저장
            # 계산된 3가지 지표를 딕셔너리 형태로 묶어 메인 결과 리스트에 추가합니다.
            results.append({
                "파일명": file_name,
                "활동성": activity_index,
                "다양성": msttr_score,
                "참여도": engagement_index
            })

        # 파일을 읽거나 분석하는 중 에러가 발생하면 멈추지 않고 에러 내용을 출력합니다.
        except Exception as e:
            print(f"❌ {file_name} 분석 중 오류: {e}")

    # ====================================================
    # 📊 최종 결과 통합 리포트 출력
    # ====================================================
    # 모인 결과 데이터를 터미널에 깔끔한 표 형태로 정렬하여 출력합니다.
    print("\n" + "="*65)
    print(f"{'파일명':<10} | {'활동성(%)':<10} | {'다양성(%)':<10} | {'참여도(%)':<10}")
    print("-" * 65)
    
    # 소수점 둘째 자리(10.2f)까지 맞춰서 출력해 가독성을 높입니다.
    for res in results:
        print(f"{res['파일명']:<12} | {res['활동성']:>10.2f} | {res['다양성']:>10.2f} | {res['참여도']:>10.2f}")
    
    print("="*65)
    print("💡 활동성: 동사/형용사 비율 | 다양성: 중복 없는 어휘 비중 | 참여도: 독자 지칭 및 청유/의문 빈도")

# 실행
# 위에서 정의한 전체 프로세스를 구동시킵니다.
analyze_multiple_newsletters()
````

## 7-6. 😊 감성 아크(Sentiment Arc) 지표 정의 및 수치화

### 🤖 감성 분석 모델

**ELECTRA (NSMC Fine-tuned)**

- 20만 건 이상의 한국어 감성 데이터(NSMC)로 학습된 모델
- 각 문장의 긍·부정 확률을 산출
- 한국어 감성 분석에 적합한 모델로 판단하여 활용

---

### 📈 감성 아크 산출 로직

뉴스레터별 길이 차이를 보정하기 위해 **구간 정규화(Normalization)** 방식을 적용하였다.

```text
전체 글
→ 진행률(0~100%) 변환
→ 10개 구간(Bin) 분할
→ 구간별 평균 감성 점수 계산
→ 감성 아크 시각화
```

> [감성 아크 시각화 이미지 삽입]

---

### 📊 감성 분석 핵심 변수

| 지표 | 정의 | 의미 |
|--------|--------|--------|
| 평균 감성 (μ) | 글 전체 감성 점수 평균 | 전반적인 정서 방향 |
| 감성 표준편차 (σ) | 감성 점수의 변동 폭 | 감정 기복 및 서사성 |
| 마지막 긍정 마감 | 마지막 10% 구간 감성 점수 | 독자의 최종 인상 및 여운 |

---

### 📌 활용 목적

- Good 글과 Bad 글의 감정 흐름 비교
- 스토리텔링 강도 분석
- 긍정적 마무리와 성과 간 관계 탐색

````python
# 딥러닝 연산을 위한 파이토치(PyTorch) 라이브러리
import torch
# 허깅페이스(Hugging Face)에서 사전 학습된 자연어 처리 모델과 토크나이저를 불러오기 위한 모듈
from transformers import AutoTokenizer, AutoModelForSequenceClassification
# 수치 연산 및 배열 처리를 위한 넘파이(NumPy)
import numpy as np
# 데이터 프레임 형태의 데이터 조작 및 분석을 위한 판다스(Pandas)
import pandas as pd
# 데이터 시각화(그래프 그리기)를 위한 맷플롯립(Matplotlib)
import matplotlib.pyplot as plt
# 정규 표현식을 사용해 문장을 분리하기 위한 모듈
import re
# 파이토치의 신경망 함수들(활성화 함수 등)을 사용하기 위한 모듈
import torch.nn.functional as F
# 파일 및 디렉토리 경로 확인 등 운영체제 기능 사용을 위한 모듈
import os

# ------------------------------------------------------------------
# 1. 감성 모델 로드
# ------------------------------------------------------------------
print("Initializing ELECTRA Sentiment Model for Long-form Text...")

# 네이버 영화 리뷰 데이터(NSMC)로 파인튜닝된 한국어 감성 분석 모델(KoELECTRA) 이름을 지정합니다.
MODEL_NAME = "daekeun-ml/koelectra-small-v3-nsmc"

# 텍스트를 모델이 이해할 수 있는 형태소(토큰) 단위로 변환해 주는 토크나이저를 불러옵니다.
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
# 긍정/부정 분류를 위해 학습된 언어 모델을 불러옵니다.
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)
# 모델을 평가(추론) 모드로 전환합니다. (드롭아웃 등 학습 시에만 쓰이는 기능 비활성화)
model.eval()


# 개별 문장을 입력받아 긍정-부정 감성 점수를 반환하는 함수입니다.
def get_sentiment_score(sentence):
    # 입력된 문장을 토큰화하고 파이토치 텐서('pt') 형태로 변환합니다.
    # 문장이 너무 길면 자르고(truncation=True), 짧으면 길이를 맞추며(padding=True), 최대 길이는 128로 제한합니다.
    inputs = tokenizer(
        sentence,
        return_tensors='pt',
        truncation=True,
        padding=True,
        max_length=128
    )

    # 기울기(gradient) 계산을 비활성화하여 메모리 사용량을 줄이고 추론 속도를 높입니다.
    with torch.no_grad():
        # 토큰화된 데이터를 모델에 통과시켜 로짓(logits, 가공되지 않은 예측값)을 얻습니다.
        outputs = model(**inputs)

    # 소프트맥스(Softmax) 함수를 적용하여 로짓 값을 0과 1 사이의 확률(probability) 값으로 변환합니다.
    probs = F.softmax(outputs.logits, dim=1)

    # 확률값 중 긍정 확률(인덱스 1)에서 부정 확률(인덱스 0)을 빼서 최종 감성 점수를 계산합니다.
    # 결과값은 -1(매우 부정)부터 1(매우 긍정) 사이의 범위를 가집니다.
    return (probs[0][1] - probs[0][0]).item()


# ------------------------------------------------------------------
# 2. 여러 파일 감성 아크 분석 (P/T 카테고리 분리 적용)
# ------------------------------------------------------------------
# 여러 텍스트 파일을 읽어 이야기 흐름(num_bins=10 구간)에 따른 감성 변화를 분석하는 함수입니다.
def analyze_multiple_files(file_list, num_bins=10):

    # P 카테고리 파일들의 요약 데이터와 그래프 데이터를 저장할 빈 변수들을 선언합니다.
    summary_rows_P = []
    plot_data_P = {}
    
    # T 카테고리 파일들의 요약 데이터와 그래프 데이터를 저장할 빈 변수들을 선언합니다.
    summary_rows_T = []
    plot_data_T = {}

    print(f"\n🚀 Starting Sentiment Arc analysis for {len(file_list)} files...\n")

    # 리스트에 담긴 파일 이름들을 하나씩 순회하며 처리합니다.
    for fname in file_list:
        file_path = f"{fname}.txt"

        # 파일이 존재하지 않으면 경고 메시지를 띄우고 다음 파일로 넘어갑니다.
        if not os.path.exists(file_path):
            print(f"⚠️ {file_path} not found. Skipping.")
            continue

        print(f"Processing {file_path}...")

        # 텍스트 파일을 UTF-8 인코딩으로 열어 전체 내용을 읽어옵니다.
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()

        # 정규 표현식을 사용해 마침표, 느낌표, 물음표 뒤의 공백이나 줄바꿈을 기준으로 텍스트를 문장 단위로 쪼갭니다.
        sentences = re.split(r'(?<=[.!?])\s+|\n', text.strip())
        # 길이가 20자 초과인 의미 있는 문장만 필터링하여 리스트로 만듭니다.
        sentences = [s.strip() for s in sentences if len(s.strip()) > 20]
        n = len(sentences)

        # 분석할 문장이 5개 미만이면 구간(bin)을 나누기 부적합하므로 건너뜁니다.
        if n < 5:
            print(f"⚠️ {fname}: 문장 수 부족")
            continue

        # 감성 점수 계산
        # 추출된 모든 문장에 대해 앞서 정의한 함수를 이용해 감성 점수를 리스트 형태로 구합니다.
        scores = [get_sentiment_score(s) for s in sentences]

        # 위치 기반 binning
        # 전체 문장의 흐름을 0부터 1 사이의 진행도(position)로 변환하여 데이터프레임에 담습니다.
        df = pd.DataFrame({
            'score': scores,
            'position': [i/(n-1) for i in range(n)]
        })

        # 문장 진행도(0~1)를 바탕으로 전체를 일정한 개수(num_bins, 기본 10개)의 구간(bin)으로 나눕니다.
        df['bin'] = pd.cut(
            df['position'],
            bins=np.linspace(0, 1, num_bins + 1),
            labels=range(1, num_bins + 1),
            include_lowest=True
        )

        # 각 구간(bin)별로 포함된 문장들의 감성 점수 평균을 계산하여 이야기 흐름에 따른 감성 변화를 도출합니다.
        bin_means = df.groupby('bin', observed=False)['score'].mean()

        # 지표 계산
        # 해당 파일의 전체 평균 감성 점수와 감성 점수의 변동폭(표준편차)을 계산합니다.
        mean_val = np.mean(scores)
        std_val = np.std(scores)
        # 이야기의 마지막 구간 점수를 확인하여 긍정적인 결말로 끝났는지 평가합니다.
        last_bin_score = bin_means.iloc[-1]
        is_positive_end = "YES" if last_bin_score > 0 else "NO"

        # 분석 결과를 딕셔너리 형태로 묶어줍니다.
        row_data = {
            "파일명": fname,
            "전체 평균 감성": mean_val,
            "감성 표준편차": std_val,
            "마지막 긍정 마감": is_positive_end
        }

        # 파일명 첫 글자를 기준으로 P 시리즈와 T 시리즈를 분류하여 각각의 리스트와 딕셔너리에 저장합니다.
        if fname.startswith("P"):
            summary_rows_P.append(row_data)
            plot_data_P[fname] = bin_means.values
        elif fname.startswith("T"):
            summary_rows_T.append(row_data)
            plot_data_T[fname] = bin_means.values
        else:
            print(f"⚠️ {fname}: P 또는 T로 시작하지 않아 분류에서 제외됩니다.")

    # ------------------------------------------------------------------
    # 3. 감성 아크 시각화 (P / T 서브플롯 분리)
    # ------------------------------------------------------------------
    # 가로 16, 세로 7 크기의 도화지에 1행 2열(좌, 우) 그래프를 생성하고, y축 눈금(sharey=True)을 공유합니다.
    fig, axes = plt.subplots(1, 2, figsize=(16, 7), sharey=True)
    # x축을 1부터 10(num_bins)까지의 구간으로 설정합니다.
    x_axis = range(1, num_bins + 1)

    # [좌측] P 카테고리 그래프
    # P 카테고리에 속한 파일들의 감성 변화 데이터를 순회하며 꺾은선 그래프로 그립니다.
    for fname, y in plot_data_P.items():
        axes[0].plot(x_axis, y, marker='o', linewidth=2, label=fname)
    # 긍정과 부정을 나누는 기준선인 0점에 검은색 점선을 긋습니다.
    axes[0].axhline(0, linestyle='--', color='black', linewidth=1)
    axes[0].set_title('Sentiment Arc Comparison - Category P', fontsize=14)
    axes[0].set_xlabel('Narrative Progress (Bins)')
    axes[0].set_ylabel('Sentiment Score (-1 ~ 1)')
    # y축의 범위를 -1.1부터 1.1로 고정합니다.
    axes[0].set_ylim(-1.1, 1.1)
    if plot_data_P:
        axes[0].legend(loc='upper left', fontsize=9)
    # 그래프에 배경 격자무늬(그리드)를 연하게 추가합니다.
    axes[0].grid(True, alpha=0.3)

    # [우측] T 카테고리 그래프
    # T 카테고리에 속한 데이터도 동일한 방식으로 우측 그래프에 그립니다.
    for fname, y in plot_data_T.items():
        axes[1].plot(x_axis, y, marker='o', linewidth=2, label=fname)
    axes[1].axhline(0, linestyle='--', color='black', linewidth=1)
    axes[1].set_title('Sentiment Arc Comparison - Category T', fontsize=14)
    axes[1].set_xlabel('Narrative Progress (Bins)')
    if plot_data_T:
        axes[1].legend(loc='upper left', fontsize=9)
    axes[1].grid(True, alpha=0.3)

    # 두 그래프가 겹치지 않게 레이아웃을 자동 조정합니다.
    plt.tight_layout()
    # 완성된 그래프를 현재 폴더에 PNG 이미지 파일로 저장합니다.
    plt.savefig('Combined_Sentiment_Arc_Categorized.png')
    # 그래프 창을 화면에 띄웁니다.
    plt.show()

    # ------------------------------------------------------------------
    # 4. 요약 표 출력 (카테고리별 분리)
    # ------------------------------------------------------------------
    # P 카테고리의 요약 데이터를 판다스 데이터프레임으로 변환하여 터미널에 텍스트 표 형태로 출력합니다.
    print("\n" + "="*70)
    print("📊 [감성 아크 요약 표 - 카테고리 P]")
    print("="*70)
    if summary_rows_P:
        df_P = pd.DataFrame(summary_rows_P)
        # 수치형 데이터는 소수점 4자리까지만 표시되도록 포맷을 지정합니다.
        print(df_P.to_string(index=False, float_format=lambda x: f"{x:.4f}"))
    else:
        print("P 카테고리 파일이 없거나 분석되지 않았습니다.")

    # T 카테고리의 요약 데이터도 동일한 방식으로 출력합니다.
    print("\n" + "="*70)
    print("📊 [감성 아크 요약 표 - 카테고리 T]")
    print("="*70)
    if summary_rows_T:
        df_T = pd.DataFrame(summary_rows_T)
        print(df_T.to_string(index=False, float_format=lambda x: f"{x:.4f}"))
    else:
        print("T 카테고리 파일이 없거나 분석되지 않았습니다.")
    print("="*70)


# ------------------------------------------------------------------
# 실행
# ------------------------------------------------------------------
# 이 파이썬 스크립트가 직접 실행될 때만 아래 블록이 동작하도록 합니다.
if __name__ == "__main__":
    target_files = [
        "P1", "P5", "P6", "P9", "P11", "P13", "P14", "P15",
        "T1", "T2", "T4", "T8", "T9", "T16", "T18", "T20"
    ]
    # 지정한 파일 목록을 바탕으로 전체 감성 분석 프로세스를 실행합니다.
    analyze_multiple_files(target_files)
````

## 7-7. 🔗 담화 응집성(Coherence) 지표 정의 및 수치화

### 🤖 담화 응집성 모델

#### 개념

텍스트 내 문장들이 논리적·의미적으로 얼마나 긴밀하게 연결되어 있는지를 측정하는 지표

#### 분석 모델

**KoBERT (Korean BERT)**

- 한국어 문맥을 이해하는 사전학습 언어모델
- 단순 단어 반복이 아닌 의미적 유사성을 기반으로 응집성 측정

---

### 📈 담화 응집성 산출 로직

```text
문장 분리
→ Sentence Embedding 생성
→ 인접 문장 간 유사도 계산
→ 전체 응집성 산출
→ 구간별 변화 추이 시각화
```

#### 측정 방법

- 각 문장을 KoBERT 임베딩 벡터로 변환
- 인접 문장(Sₙ, Sₙ₊₁) 간 코사인 유사도 계산
- 전체 평균값을 담화 응집성 지표로 활용

```text
Cosine Similarity

= (A · B) / (||A|| ||B||)
```

> [담화 응집성 시각화 이미지 삽입]

---

### 📊 담화 응집성 핵심 변수

| 지표 | 정의 | 의미 |
|------|------|------|
| 평균 응집성 | 문장 간 평균 코사인 유사도 | 글의 전반적인 논리성 |
| 응집성 흐름 | 구간별 응집성 변화 추이 | 전개 과정의 연결성 변화 |

---

### 📌 활용 목적

- Good 글과 Bad 글의 논리적 연결성 비교
- 글 전개 방식 및 구조적 완성도 분석
- 카테고리별 서술 패턴 차이 탐색

````python
# 딥러닝 모델 구동을 위한 파이토치
import torch
# 수치 및 배열 계산을 위한 넘파이
import numpy as np
# 데이터 정렬 및 표 생성을 위한 판다스
import pandas as pd
# 데이터 시각화(그래프)를 위한 맷플롯립
import matplotlib.pyplot as plt
# 정규 표현식을 이용한 문장 분리를 위해 불러옴
import re
# 파일 존재 여부 확인 등 OS 제어를 위해 불러옴
import os
# 코사인 유사도 등 파이토치의 수학적 함수를 사용하기 위해 불러옴
import torch.nn.functional as F
# 허깅페이스의 BERT 토크나이저와 모델을 불러오기 위한 모듈
from transformers import BertTokenizer, BertModel

# ------------------------------------------------------------------
# 1. 모델 로드 (KoBERT)
# ------------------------------------------------------------------
print("Initializing KoBERT for Coherence Stability Analysis...")

# SKT에서 만든 한국어 전용 BERT 모델(KoBERT)의 이름을 지정합니다.
BERT_NAME = "skt/kobert-base-v1"

# 텍스트를 모델이 이해할 수 있는 단위로 쪼개주는 토크나이저를 불러옵니다.
tokenizer = BertTokenizer.from_pretrained(BERT_NAME)
# 문장 벡터(임베딩)를 추출하기 위해 사전 학습된 KoBERT 모델을 불러옵니다.
model = BertModel.from_pretrained(BERT_NAME)
# 모델을 평가 모드로 설정하여, 학습 시에만 쓰이는 기능들을 끕니다.
model.eval()

# ------------------------------------------------------------------
# 2. 분석 함수
# ------------------------------------------------------------------
# 단일 문장을 입력받아 모델을 통해 다차원 벡터(임베딩)로 변환하는 함수입니다.
def get_embedding(sentence):
    # 문장을 토큰화하고 파이토치 텐서 형태로 변환합니다. 최대 길이는 128로 맞춥니다.
    inputs = tokenizer(
        sentence, return_tensors='pt',
        truncation=True, padding=True, max_length=128
    )
    # 기울기 계산을 비활성화하여 추론 속도를 높이고 메모리를 절약합니다.
    with torch.no_grad():
        outputs = model(**inputs)
    # 모델의 출력 중 [CLS] 토큰에 해당하는 첫 번째 토큰의 벡터(문장 전체의 의미를 담음)만 추출하여 반환합니다.
    return outputs.last_hidden_state[:, 0, :]

# 파일 경로를 받아 문장 간의 의미적 연결성(응집성) 점수를 리스트로 반환하는 함수입니다.
def calculate_coherence(file_path):
    # 파일이 없으면 분석을 중단하고 None을 반환합니다.
    if not os.path.exists(file_path):
        return None

    # 파일을 읽고, 줄바꿈 기호를 띄어쓰기로 바꿔 하나의 긴 텍스트로 만듭니다.
    with open(file_path, 'r', encoding='utf-8') as f:
        text = f.read().replace('\n', ' ')
    
    # 마침표, 느낌표, 물음표를 기준으로 텍스트를 문장 단위로 분리합니다.
    sentences = re.split(r'(?<=[.!?])\s+', text.strip())
    # 길이가 5글자 초과인 유의미한 문장만 남깁니다.
    sentences = [s for s in sentences if len(s.strip()) > 5]
    n = len(sentences)

    # 문장이 2개 미만이면 비교할 대상이 없으므로 None을 반환합니다.
    if n < 2:
        return None

    # 문장 간 유사도 점수를 담을 빈 리스트를 만듭니다.
    coherence_scores = []
    # 첫 번째 문장의 벡터값을 먼저 구해놓습니다.
    prev_emb = get_embedding(sentences[0])

    # 두 번째 문장부터 마지막 문장까지 순회하며 바로 앞 문장과의 유사도를 계산합니다.
    for i in range(1, n):
        # 현재 문장의 벡터값을 구합니다.
        curr_emb = get_embedding(sentences[i])
        # 앞 문장과 현재 문장 벡터 간의 코사인 유사도(방향의 일치 정도)를 계산해 실수값으로 뽑아냅니다.
        sim = F.cosine_similarity(prev_emb, curr_emb).item()
        # 계산된 유사도를 리스트에 추가합니다.
        coherence_scores.append(sim)
        # 다음 루프를 위해 현재 문장을 '이전 문장'으로 업데이트합니다.
        prev_emb = curr_emb

    return coherence_scores

# ------------------------------------------------------------------
# 3. 메인 분석 및 시각화 (P/T 카테고리 분리 적용)
# ------------------------------------------------------------------
# 여러 파일의 응집성 안정성을 분석하고 시각화하는 메인 함수입니다.
def analyze_coherence_stability(file_list, num_bins=10):
    # P/T 카테고리별 요약 데이터 저장을 위한 빈 리스트를 준비합니다.
    summary_results_P = []
    summary_results_T = []
    
    # P/T 카테고리별 그래프 그리기용 데이터를 담을 빈 리스트를 준비합니다.
    plot_data_P = []
    plot_data_T = []

    # 주어진 파일 리스트를 하나씩 순회합니다.
    for file_name in file_list:
        file_path = f"{file_name}.txt"
        # 위에서 만든 함수를 통해 해당 파일의 문장 간 응집성 점수들을 구합니다.
        scores = calculate_coherence(file_path)
        
        # 파일이 없거나 문장이 부족해 None이 반환되었다면 건너뜁니다.
        if scores is None: continue

        # 통계 계산
        # 전체 문장 간 유사도의 평균을 구합니다.
        avg_coh = np.mean(scores)
        # 유사도의 표준편차를 구합니다. (작을수록 흐름이 튀지 않고 안정적임)
        std_coh = np.std(scores) # 응집성 표준편차
        
        # 행 데이터를 딕셔너리로 만듭니다.
        row_data = {
            "파일명": file_name,
            "평균 응집성": avg_coh,
            "응집성 표준편차": std_coh  # 연결의 일관성 지표
        }

        # 구간별 시각화 (흐름 확인) 데이터 가공
        # 전체 이야기 흐름을 0.0부터 1.0 사이로 정규화합니다.
        x_norm = np.linspace(0, 1, len(scores))
        # 지정된 구간(기본 10구간)으로 나누기 위한 기준점들을 만듭니다.
        bins = np.linspace(0, 1, num_bins + 1)
        # 각 점수가 몇 번째 구간에 속하는지 인덱스를 부여합니다.
        bin_indices = np.digitize(x_norm, bins[1:-1])
        
        # 빈 구간 처리 (RuntimeWarning 방지)
        bin_means = []
        # 총 10개의 구간을 순회하며 각 구간에 속한 유사도 점수들의 평균을 구합니다.
        for i in range(num_bins):
            # 현재 구간(i)에 해당하는 점수들만 필터링합니다.
            scores_in_bin = np.array(scores)[bin_indices == i]
            # 해당 구간에 점수가 있으면 평균을, 없으면 0을 리스트에 추가합니다.
            bin_means.append(np.mean(scores_in_bin) if len(scores_in_bin) > 0 else 0)
        
        # 파일명 첫 글자를 기준으로 카테고리를 P와 T로 분류하여 각각 저장합니다.
        if file_name.startswith("P"):
            summary_results_P.append(row_data)
            # 그래프용 데이터는 (파일명, 구간별 평균점수, 표준편차) 형태로 묶습니다.
            plot_data_P.append((file_name, bin_means, std_coh))
        elif file_name.startswith("T"):
            summary_results_T.append(row_data)
            plot_data_T.append((file_name, bin_means, std_coh))
        else:
            print(f"⚠️ {file_name}: P 또는 T로 시작하지 않아 분류에서 제외됩니다.")

    # ------------------------------------------------------------------
    # 그래프 출력 (P / T 서브플롯 분리)
    # ------------------------------------------------------------------
    # 가로 16, 세로 7 크기의 1행 2열 서브플롯을 만들고, y축 범위를 동일하게 공유합니다.
    fig, axes = plt.subplots(1, 2, figsize=(16, 7), sharey=True)
    # x축을 1부터 지정된 구간 수(10)까지로 설정합니다.
    x_axis = range(1, num_bins + 1)

    # [좌측] P 카테고리 그래프
    # P 카테고리 데이터를 하나씩 꺼내어 꺾은선 그래프로 시각화합니다.
    for name, data, std in plot_data_P:
        axes[0].plot(x_axis, data, marker='o', label=f"{name} (σ={std:.3f})")
    # 그래프 제목, 축 이름, 배경 격자를 설정합니다.
    axes[0].set_title('Coherence Consistency - Category P', fontsize=14)
    axes[0].set_xlabel('Narrative Progress (Bins)', fontsize=12)
    axes[0].set_ylabel('Semantic Coherence', fontsize=12)
    axes[0].grid(True, linestyle='--', alpha=0.5)
    # 데이터가 있다면 범례를 그립니다.
    if plot_data_P:
        axes[0].legend(title="Files (σ = Std Dev)", loc='upper left', fontsize=9)

    # [우측] T 카테고리 그래프
    # T 카테고리 데이터도 동일한 방식으로 우측 화면에 그립니다.
    for name, data, std in plot_data_T:
        axes[1].plot(x_axis, data, marker='o', label=f"{name} (σ={std:.3f})")
    axes[1].set_title('Coherence Consistency - Category T', fontsize=14)
    axes[1].set_xlabel('Narrative Progress (Bins)', fontsize=12)
    axes[1].grid(True, linestyle='--', alpha=0.5)
    if plot_data_T:
        axes[1].legend(title="Files (σ = Std Dev)", loc='upper left', fontsize=9)

    # 텍스트와 그래프가 겹치지 않게 여백을 자동 조절한 뒤, 이미지 파일로 저장하고 화면에 띄웁니다.
    plt.tight_layout()
    plt.savefig('Combined_Coherence_Stability_Categorized.png')
    plt.show()

    # ------------------------------------------------------------------
    # 결과 테이블 출력 (카테고리별 분리)
    # ------------------------------------------------------------------
    # P 카테고리의 결과 데이터를 표 형태로 터미널에 출력합니다.
    print("\n" + "="*60)
    print("📊 [응집성 안정성 요약 표 - 카테고리 P]")
    print("="*60)
    if summary_results_P:
        # 데이터프레임으로 변환 후, 표준편차가 작은 순(일관성이 높은 순)으로 정렬해 출력합니다.
        df_P = pd.DataFrame(summary_results_P)
        print(df_P.sort_values(by="응집성 표준편차").to_string(index=False, float_format=lambda x: f"{x:.4f}"))
    else:
        print("P 카테고리 파일이 없거나 분석되지 않았습니다.")
        
    # T 카테고리의 결과 데이터도 동일한 기준으로 표를 생성하여 출력합니다.
    print("\n" + "="*60)
    print("📊 [응집성 안정성 요약 표 - 카테고리 T]")
    print("="*60)
    if summary_results_T:
        df_T = pd.DataFrame(summary_results_T)
        print(df_T.sort_values(by="응집성 표준편차").to_string(index=False, float_format=lambda x: f"{x:.4f}"))
    else:
        print("T 카테고리 파일이 없거나 분석되지 않았습니다.")
    print("="*60)
    print("* 표준편차가 작을수록 문장 간 연결이 일관되고 매끄러운 담화입니다.")

# ------------------------------------------------------------------
# 실행
# ------------------------------------------------------------------
# 해당 파이썬 파일이 메인으로 실행될 때만 아래 분석을 동작하게 합니다.
if __name__ == "__main__":
    target_files = [
        "P1", "P5", "P6", "P9", "P11", "P13", "P14", "P15",
        "T1", "T2", "T4", "T8", "T9", "T16", "T18", "T20"
    ]
    analyze_coherence_stability(target_files)
````

# 8. 📈 Good·Bad 그룹 평균 비교 및 핵심 지표 선정 기준

## 개요

7장에서 정의한 텍스트 지표를 활용하여 Good 그룹과 Bad 그룹 간 차이를 비교 분석하였다.

본 분석의 목적은 단순히 통계적 차이를 확인하는 것이 아니라, 실제 뉴스레터 성과와 연관된 텍스트 특성을 발견하고 향후 뉴스레터 작성 시 활용 가능한 실무적 가이드라인을 도출하는 데 있다.

---

## 8-1. 📊 그룹별 평균 비교

각 지표에 대해 Good 그룹과 Bad 그룹의 평균값을 비교하였다.

평균 비교를 수행한 이유는 성과가 높은 뉴스레터가 공통적으로 보이는 특징을 정량적으로 확인하고, 향후 콘텐츠 작성 시 참고할 수 있는 기준값(Target Baseline)을 도출하기 위함이다.

예를 들어 Good 그룹의 참여도가 일관되게 높게 나타난다면, 이는 독자와의 상호작용을 강화하는 문체가 성과와 관련될 가능성을 시사한다.

---

## 8-2. 📉 p-value(유의확률) 해석 기준

평균 차이가 우연히 발생했을 가능성을 확인하기 위해 p-value를 함께 산출하였다.

다만 본 프로젝트는 총 16개의 뉴스레터를 분석 대상으로 하는 소표본 연구이므로, 일반적인 학술 연구에서 사용하는 엄격한 유의수준(p < 0.05)을 절대적인 기준으로 적용하는 데 한계가 있다.

따라서 본 분석에서는 p-value를 참고 지표로 활용하되,

> **통계적 유의성보다 실무적 활용 가능성과 일관된 패턴 발견에 더 큰 의미를 두었다.**

---

## 8-3. 🎯 효과크기(Effect Size) 중심 해석

본 프로젝트에서는 p-value보다 효과크기(Cohen's d)를 핵심 판단 기준으로 활용하였다.

p-value가 차이의 존재 가능성을 보여주는 지표라면, 효과크기는 해당 차이가 실제로 얼마나 큰 영향력을 가지는지를 설명한다.

즉,

* p-value → 차이가 존재하는가?
* Effect Size → 차이가 얼마나 중요한가?

를 의미한다.

실제 콘텐츠 제작 환경에서는 통계적 유의성보다 성과 개선에 직접 연결되는 영향력이 더욱 중요하므로, 효과크기를 중심으로 결과를 해석하였다.

---

## 8-4. ✅ 핵심 지표 선정 기준

효과크기(Cohen's d)의 절댓값을 기준으로 핵심 지표를 선별하였다.

| Cohen's d | 해석                 |
| --------- | ------------------ |
| 0.2 이상    | 작은 효과              |
| 0.5 이상    | 중간 효과              |
| 0.8 이상    | 큰 효과(Large Effect) |

본 프로젝트에서는 다음 기준을 적용하였다.

> **효과크기 절댓값이 0.8 이상(|d| ≥ 0.8)인 경우를 핵심 지표로 선정**

이는 Good 그룹과 Bad 그룹 간 차이가 실질적으로 크다고 판단되는 수준이며, 향후 자동 평가 시스템의 피드백 기준(Threshold)으로 활용할 수 있는 후보 지표로 간주하였다.

---

## 📌 분석 활용 방향

최종적으로 각 텍스트 지표에 대해

* Good·Bad 그룹 평균 비교
* p-value 검정
* 효과크기(Cohen's d) 산출

을 수행하였으며,

효과크기가 크게 나타난 지표를 중심으로 뉴스레터 성과와 연관된 핵심 텍스트 특성을 도출하고자 하였다.

이후 장에서는 각 카테고리별 분석 결과를 바탕으로 Good 콘텐츠와 Bad 콘텐츠의 차이를 비교하고, 실제 글쓰기 가이드라인으로 활용 가능한 인사이트를 제시한다.


````python
# 데이터 조작 및 표 형태의 데이터 분석을 위한 판다스(Pandas)
import pandas as pd
# 수치 계산과 배열 연산을 위한 넘파이(NumPy)
import numpy as np
# T-test 등 통계 검정 기능을 사용하기 위해 사이파이(SciPy)의 stats 모듈을 불러옵니다.
from scipy import stats

def calculate_cohens_d(group1, group2):
    """Cohen's d (효과크기) 계산 함수"""
    # 두 집단의 데이터 개수(n1, n2)를 구합니다.
    n1, n2 = len(group1), len(group2)
    # ddof=1을 설정하여 표본의 분산(Variance)을 계산합니다.
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    
    # 통합 표준편차(Pooled Standard Deviation) 계산 공식입니다.
    # 두 집단의 분산을 각각의 자유도(n-1)로 가중 평균한 뒤 제곱근을 씌웁니다.
    pooled_std = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    
    # 분모가 0이 되어 에러가 발생하는 것을 방지합니다.
    if pooled_std == 0:
        return 0
    # 두 집단의 평균 차이를 통합 표준편차로 나누어 최종 효과크기(Cohen's d)를 반환합니다.
    return (np.mean(group1) - np.mean(group2)) / pooled_std

def analyze_statistics(df):
    # 1. Good / Bad 기준 정의
    # 기존 성과를 바탕으로 우수(Good)와 부진(Bad) 뉴스레터를 파일명 기준으로 사전에 정의합니다.
    p_good = ['P1', 'P4', 'P9', 'P15']
    p_bad = ['P5', 'P6', 'P13', 'P14']
    t_good = ['T1', 'T2', 'T4', 'T8']
    t_bad = ['T9', 'T16', 'T18', 'T20']

    # 2. 분류 파생 변수 생성
    # 파일명의 첫 글자를 확인하여 'P'로 시작하면 Pilgrimage, 'T'로 시작하면 The Trail 카테고리로 묶어주는 함수입니다.
    def get_category(filename):
        if filename.startswith('P'): return 'Pilgrimage (P)'
        elif filename.startswith('T'): return 'The Trail (T)'
        return 'Unknown'

    # 파일명이 앞서 정의한 Good 리스트에 있는지 Bad 리스트에 있는지 판별하여 그룹을 묶어주는 함수입니다.
    def get_group(filename):
        if filename in p_good or filename in t_good: return 'Good'
        elif filename in p_bad or filename in t_bad: return 'Bad'
        return 'Unknown'

    # 원본 데이터프레임(df)에 분류 함수들을 적용해 '카테고리'와 '그룹'이라는 새로운 열(파생 변수)을 만듭니다.
    df['카테고리'] = df['파일명'].apply(get_category)
    df['그룹'] = df['파일명'].apply(get_group)
    
    # 평가 대상(Good/Bad)이 아닌 데이터('Unknown')는 분석에서 제외한 새로운 데이터프레임을 만듭니다.
    df_filtered = df[df['그룹'] != 'Unknown']

    # 분석할 지표(숫자형 컬럼) 목록을 지정합니다.
    numeric_cols = ['활동성(%)', '다양성(%)', '참여도(%)', '전체 평균 감성', '감성 표준편차', '평균 응집성', '응집성 표준편차']
    
    # 검정 결과를 차곡차곡 담을 빈 리스트를 생성합니다.
    results = []

    # 3. 카테고리별 통계 검정 수행
    # 두 가지 카테고리(P, T)에 대해 각각 반복문을 실행합니다.
    for category in ['Pilgrimage (P)', 'The Trail (T)']:
        # 현재 반복 중인 카테고리에 해당하는 데이터만 추출합니다.
        cat_df = df_filtered[df_filtered['카테고리'] == category]
        
        # 각 숫자형 지표(컬럼)별로 반복문을 실행합니다.
        for col in numeric_cols:
            # 해당 카테고리와 지표에서 Good 그룹의 수치들만 배열로 추출합니다.
            good_vals = cat_df[cat_df['그룹'] == 'Good'][col].values
            # 동일하게 Bad 그룹의 수치들만 추출합니다.
            bad_vals = cat_df[cat_df['그룹'] == 'Bad'][col].values
            
            # Welch's t-test (등분산성을 가정하지 않음)
            # 두 그룹의 데이터 분산이 다를 수 있다고 가정(equal_var=False)하고 평균 차이가 통계적으로 유의미한지 검정합니다.
            t_stat, p_val = stats.ttest_ind(good_vals, bad_vals, equal_var=False)
            
            # 효과크기 (Cohen's d)
            # 통계적 유의미성(p-value) 외에, 실제 평균 차이가 얼마나 큰지 실질적 의미(효과크기)를 구합니다.
            effect_size = calculate_cohens_d(good_vals, bad_vals)
            
            # 산출된 통계 정보들을 딕셔너리 형태로 묶어 결과 리스트에 추가합니다.
            results.append({
                '카테고리': category,
                '평가 지표': col,
                'Good 평균': np.mean(good_vals),
                'Bad 평균': np.mean(bad_vals),
                'p-value': p_val,
                '효과크기(Cohen\'s d)': effect_size
            })

    # 4. 데이터프레임 변환 및 출력 포맷팅
    # 딕셔너리가 담긴 결과 리스트를 판다스 데이터프레임으로 변환합니다.
    res_df = pd.DataFrame(results)
    
    # 유의미한 결과(p < 0.05 또는 |d| > 0.8)에 강조 표시를 위한 함수
    # p-value가 0.05 미만이면 통계적으로 유의미하므로 뒤에 '*' 기호를 붙입니다.
    def format_p(val):
        return f"{val:.4f} *" if val < 0.05 else f"{val:.4f}"
        
    # 효과크기(Cohen's d)의 절대값이 0.8 이상이면 매우 큰 차이를 의미하므로 '💡' 기호를 붙입니다.
    def format_d(val):
        return f"{val:.3f} 💡" if abs(val) >= 0.8 else f"{val:.3f}"

    # 가독성을 높이기 위해 수치 데이터를 소수점 셋째 자리까지만 표시하도록 반올림합니다.
    res_df['Good 평균'] = res_df['Good 평균'].round(3)
    res_df['Bad 평균'] = res_df['Bad 평균'].round(3)
    # 위에서 정의한 포맷팅 함수들을 각 열에 적용하여 텍스트 형태를 수정합니다.
    res_df['p-value (유의확률)'] = res_df['p-value'].apply(format_p)
    res_df['효과크기 (Cohen\'s d)'] = res_df['효과크기(Cohen\'s d)'].apply(format_d)
    
    # 보기 좋게 정리
    # 최종 출력에 필요한 컬럼들만 순서대로 골라 새로운 데이터프레임을 만듭니다.
    final_df = res_df[['카테고리', '평가 지표', 'Good 평균', 'Bad 평균', 'p-value (유의확률)', '효과크기 (Cohen\'s d)']]

    # 결과를 터미널에 출력하기 위한 안내 문구를 작성합니다.
    print("\n" + "="*90)
    print("🔬 [카테고리별 Good vs Bad 그룹 통계 분석 결과]")
    print("  * p-value < 0.05 인 경우 '*' 표시 (통계적 유의미)")
    print("  💡 효과크기 절대값이 0.8 이상인 경우 '💡' 표시 (실질적으로 큰 차이)")
    print("="*90)
    
    # P와 T 카테고리별로 분리해서 표(table) 형태로 출력합니다.
    for cat in ['Pilgrimage (P)', 'The Trail (T)']:
        print(f"\n[{cat}]")
        # 현재 출력하려는 카테고리만 남기고, 이미 제목에 표기했으므로 '카테고리' 컬럼은 제거(drop)합니다.
        display_df = final_df[final_df['카테고리'] == cat].drop(columns=['카테고리'])
        # 인덱스 번호를 지우고 깔끔한 문자열 표 형태로 출력합니다.
        print(display_df.to_string(index=False))
        print("-" * 80)

# ------------------------------------------------------------------
# 실행부
# ------------------------------------------------------------------
# 파이썬 스크립트가 메인으로 직접 실행될 때 작동하는 구역입니다.
if __name__ == "__main__":
    
    # 제공해주신 전체 16개 파일의 모든 분석 수치를 명시한 원본 데이터 딕셔너리입니다.
    full_data = {
        '파일명': [
            'P1', 'P5', 'P6', 'P9', 'P11', 'P13', 'P14', 'P15',
            'T1', 'T2', 'T4', 'T8', 'T9', 'T16', 'T18', 'T20'
        ],
        '활동성(%)': [
            6.85, 7.11, 8.00, 11.54, 8.99, 9.66, 7.97, 8.60,
            12.04, 10.24, 10.78, 13.18, 11.45, 11.50, 10.89, 10.37
        ],
        '다양성(%)': [
            41.87, 42.67, 44.53, 45.10, 45.20, 46.40, 46.80, 46.40,
            42.47, 45.20, 42.60, 44.00, 43.80, 46.80, 48.20, 48.60
        ],
        '참여도(%)': [
            0.56, 0.19, 0.42, 0.43, 0.82, 0.62, 0.76, 0.89,
            0.19, 1.12, 0.97, 1.21, 1.42, 1.45, 1.15, 0.83
        ],
        '전체 평균 감성': [
            0.5537, 0.4190, 0.7293, 0.2543, 0.2284, 0.2410, 0.1627, 0.4294,
            0.3501, 0.6201, 0.5302, 0.5507, 0.3719, 0.1948, 0.5120, 0.4101
        ],
        '감성 표준편차': [
            0.7399, 0.8154, 0.6247, 0.9087, 0.8427, 0.8600, 0.8487, 0.8152,
            0.8716, 0.6771, 0.7729, 0.8142, 0.8585, 0.8490, 0.7422, 0.8659
        ],
        '평균 응집성': [
            0.9944, 0.9155, 0.9017, 0.8488, 0.8198, 0.9097, 0.9124, 0.9056,
            0.8810, 0.9515, 0.8755, 0.7845, 0.8620, 0.8954, 0.9770, 0.9076
        ],
        '응집성 표준편차': [
            0.0075, 0.1674, 0.1749, 0.2347, 0.2762, 0.1718, 0.1673, 0.1745,
            0.2301, 0.1294, 0.2828, 0.3104, 0.2736, 0.2480, 0.0747, 0.2494
        ]
    }
    
    # 딕셔너리 데이터를 판다스 데이터프레임 구조로 변환합니다.
    df = pd.DataFrame(full_data)
    # 만들어진 데이터프레임을 통계 분석 함수에 넣어 프로그램을 실행합니다.
    analyze_statistics(df)
````

## 9-3. 🔄 분석 방향 재설정: 다회오픈율 기반 성과 재평가

Pilgrimage 카테고리 분석 결과를 통해, 텍스트 지표 전체에서 효과크기가 낮게 나타나 Good 그룹과 Bad 그룹 간의 뚜렷한 차이를 확인하기 어려웠다. 이는 텍스트 품질 자체의 문제라기보다, 현재 성과 분류 기준인 오픈율(Open Rate)과 CTOR이 콘텐츠의 본질적 가치를 충분히 반영하지 못하고 있을 가능성을 시사한다.

오픈율과 CTOR은 제목, 인물 및 브랜드의 인지도, 애디터 콘텐츠 내 클릭 지표 활용 실패 등 다양한 외부 요인의 영향을 받을 수 있다. 즉, 콘텐츠를 클릭하게 만드는 힘은 측정할 수 있지만, 콘텐츠 자체가 얼마나 가치 있게 소비되었는지는 충분히 설명하기 어렵다.

이에 따라 본 프로젝트의 핵심 문제 정의인 **"좋은 글이란 무엇인가?"** 를 재설정하고자 한다. 즉, The Trail 성과 분류 기준으로 활용 중인, 독자의 반복 소비를 반영하는 **다회오픈율(Multiple Open Rate)** 을 활용하였다.

```text
다회오픈율
=
2회 이상 오픈한 구독자 수
÷
전체 오픈 구독자 수
```

다회오픈율은 단순 클릭이나 최초 열람이 아닌, 독자가 콘텐츠를 다시 읽고 싶을 만큼 몰입했는지를 보여주는 지표이다. 따라서 이후 분석에서는 Pilgrimage 카테고리 역시 다회오픈율을 기준으로 Good·Bad 그룹을 재구성하여 재분석을 수행하고, 콘텐츠 자체의 힘(Content Quality)과 연관된 텍스트 특성을 탐색하고자 하였다.

# 10. 📈 다회오픈율 기반 Pilgrimage 텍스트 데이터 분석 결과 및 시사점

## 개요

9장에서 다회오픈율을 기반으로 Pilgrimage 카테고리의 Good·Bad 그룹을 재구성한 후 텍스트 지표를 비교 분석하였다.

그 결과, 기존 오픈율·CTOR 기준 분석에서는 확인되지 않았던 여러 지표에서 큰 효과크기(|d| ≥ 0.8)가 나타났으며, 이는 콘텐츠의 반복 소비와 관련된 텍스트 특성을 보여주는 의미 있는 신호로 해석할 수 있다.

---

````python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv('뉴스레터_통계.csv', encoding='cp949')

# The Trail만 추출
trail = df[df['카테고리'] == 'Pilgrimage']

# 상위 4개
top4 = trail.nlargest(4, '다회오픈율')

# 하위 4개
bottom4 = trail.nsmallest(4, '다회오픈율')

print("=== 다회오픈율 상위 4개 ===")
print(top4[['Index', '다회오픈율']])

print("\n=== 다회오픈율 하위 4개 ===")
print(bottom4[['Index', '다회오픈율']])
````

````python
import pandas as pd
import numpy as np
from scipy import stats

def calculate_cohens_d(group1, group2):
    """Cohen's d (효과크기) 계산 함수"""
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    
    # 통합 표준편차(Pooled Standard Deviation)
    pooled_std = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    
    if pooled_std == 0:
        return 0
    return (np.mean(group1) - np.mean(group2)) / pooled_std

def analyze_statistics(df):
    # 1. Pilgrimage (P) 그룹 기준 정의
    p_good = ['P1', 'P2', 'P4', 'P5']
    p_bad = ['P9', 'P10', 'P11', 'P16']

    # 2. 분류 파생 변수 생성
    def get_group(filename):
        if filename in p_good: return 'Good'
        elif filename in p_bad: return 'Bad'
        return 'Unknown'

    df['그룹'] = df['파일명'].apply(get_group)
    df_filtered = df[df['그룹'] != 'Unknown']

    # 분석할 지표 목록
    numeric_cols = ['활동성(%)', '다양성(%)', '참여도(%)', '전체 평균 감성', '감성 표준편차', '평균 응집성', '응집성 표준편차']
    
    results = []

    # 3. Pilgrimage 카테고리 통계 검정 수행
    cat_df = df_filtered
    for col in numeric_cols:
        good_vals = cat_df[cat_df['그룹'] == 'Good'][col].values
        bad_vals = cat_df[cat_df['그룹'] == 'Bad'][col].values
        
        # Welch's t-test
        t_stat, p_val = stats.ttest_ind(good_vals, bad_vals, equal_var=False)
        
        # 효과크기 (Cohen's d)
        effect_size = calculate_cohens_d(good_vals, bad_vals)
        
        results.append({
            '평가 지표': col,
            'Good 평균': np.mean(good_vals),
            'Bad 평균': np.mean(bad_vals),
            'p-value': p_val,
            '효과크기(Cohen\'s d)': effect_size
        })

    # 4. 데이터프레임 변환 및 출력 포맷팅
    res_df = pd.DataFrame(results)
    
    def format_p(val):
        return f"{val:.4f} *" if val < 0.05 else f"{val:.4f}"
        
    def format_d(val):
        return f"{val:.3f} 💡" if abs(val) >= 0.8 else f"{val:.3f}"

    res_df['Good 평균'] = res_df['Good 평균'].round(3)
    res_df['Bad 평균'] = res_df['Bad 평균'].round(3)
    res_df['p-value (유의확률)'] = res_df['p-value'].apply(format_p)
    res_df['효과크기 (Cohen\'s d)'] = res_df['효과크기(Cohen\'s d)'].apply(format_d)
    
    # 보기 좋게 정리
    final_df = res_df[['평가 지표', 'Good 평균', 'Bad 평균', 'p-value (유의확률)', '효과크기 (Cohen\'s d)']]

    print("\n" + "="*80)
    print("🔬 [Pilgrimage (P) Good vs Bad 그룹 통계 분석 결과]")
    print("  * p-value < 0.05 인 경우 '*' 표시")
    print("  💡 효과크기 절대값이 0.8 이상인 경우 '💡' 표시")
    print("="*80)
    print(final_df.to_string(index=False))
    print("-" * 80)

# ------------------------------------------------------------------
# 실행부
# ------------------------------------------------------------------
if __name__ == "__main__":
    full_data = {
        '파일명': ['P1', 'P2', 'P4', 'P5', 'P9', 'P10', 'P11', 'P16'],
        '활동성(%)': [6.85, 7.39, 8.05, 7.11, 11.54, 10.15, 8.99, 7.54],
        '다양성(%)': [41.87, 42.60, 45.53, 42.67, 45.10, 45.30, 45.20, 44.80],
        '참여도(%)': [0.56, 2.14, 0.22, 0.19, 0.43, 0.58, 0.82, 0.60],
        '전체 평균 감성': [0.5537, 0.6568, 0.2555, 0.4190, 0.2543, 0.1969, 0.2284, 0.4274],
        '감성 표준편차': [0.7399, 0.6708, 0.8945, 0.8154, 0.9087, 0.8663, 0.8427, 0.7674],
        '평균 응집성': [0.9944, 0.8081, 0.8619, 0.9155, 0.8488, 0.8390, 0.8198, 0.9522],
        '응집성 표준편차': [0.0075, 0.2922, 0.2749, 0.1674, 0.2347, 0.2531, 0.2762, 0.1177]
    }
    df = pd.DataFrame(full_data)
    analyze_statistics(df)
````

## 📌 프로젝트 최종 성과

본 프로젝트는 단순히 뉴스레터 성과를 분석하는 데 그치지 않고,

이를 통해 에디터는 발행 전 뉴스레터의 텍스트 특성을 정량적으로 점검할 수 있으며, 데이터 기반의 글쓰기 환경을 구축할 수 있는 기반을 마련하였다.

````python
# 파일 경로 탐색 및 생성을 위한 운영체제 모듈
import os
# 문장 분리 등 정규 표현식을 사용하기 위한 모듈
import re
# 평균, 표준편차 등 수치 연산을 위한 넘파이
import numpy as np
# 딥러닝 모델 구동을 위한 파이토치
import torch
# 소프트맥스 등 파이토치 신경망 함수들을 위한 모듈
import torch.nn.functional as F
# 한국어 형태소 분석기 키위(Kiwi)
from kiwipiepy import Kiwi
# 허깅페이스의 사전 학습 모델과 토크나이저를 불러오기 위한 모듈
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# ------------------------------------------------------------------
# 1. 모델 초기화 (형태소 분석기 및 감성 모델)
# ------------------------------------------------------------------
print("시스템 로딩 중... (Kiwi 및 ELECTRA 모델 초기화)")
# 형태소 분석기 객체를 생성합니다.
kiwi = Kiwi()

# 한국어 감성 분석에 특화된 KoELECTRA 모델 이름을 지정합니다.
MODEL_NAME = "daekeun-ml/koelectra-small-v3-nsmc"
# 텍스트를 모델이 이해할 수 있는 토큰으로 변환하는 토크나이저를 로드합니다.
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
# 긍정/부정 분류용 사전 학습 모델을 로드합니다.
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)
# 모델을 추론(평가) 모드로 설정하여 불필요한 연산을 막습니다.
model.eval()

# 단일 문장을 입력받아 긍정-부정 감성 점수(-1 ~ 1)를 반환하는 함수입니다.
def get_sentiment_score(sentence):
    # 문장을 파이토치 텐서 형태로 토큰화하고 길이를 맞춥니다.
    inputs = tokenizer(sentence, return_tensors='pt', truncation=True, padding=True, max_length=128)
    # 기울기 계산을 비활성화하여 추론 속도를 높입니다.
    with torch.no_grad():
        outputs = model(**inputs)
    # 로짓(예측값)을 0~1 사이의 확률값으로 변환합니다.
    probs = F.softmax(outputs.logits, dim=1)
    # 긍정 확률(1)에서 부정 확률(0)을 뺀 값을 반환합니다.
    return (probs[0][1] - probs[0][0]).item()

# ------------------------------------------------------------------
# 2. 텍스트 분석 및 피드백 리포트 생성 함수
# ------------------------------------------------------------------
# 파일 경로와 카테고리(P 또는 T)를 입력받아 분석 리포트를 출력하는 함수입니다.
def evaluate_new_newsletter(file_path, category):
    # 파일이 존재하지 않으면 경고를 출력하고 함수를 종료합니다.
    if not os.path.exists(file_path):
        print(f"⚠️ 파일을 찾을 수 없습니다: {file_path}")
        return

    # 텍스트 파일을 읽기 모드로 엽니다.
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()

    # 텍스트 전처리
    # 분석에 방해되는 "첨부 이미지" 단어를 제거하고 양끝 공백을 지웁니다.
    clean_text = content.replace("첨부 이미지", "").strip()
    # 마침표, 느낌표, 물음표 또는 줄바꿈을 기준으로 텍스트를 문장 단위로 쪼갭니다.
    sentences = re.split(r'(?<=[.!?])\s+|\n', clean_text)
    # 길이가 20자 초과인 의미 있는 문장만 필터링하여 리스트로 만듭니다.
    sentences = [s.strip() for s in sentences if len(s.strip()) > 20]
    
    # 분석할 문장이 없으면 경고 후 종료합니다.
    if not sentences:
        print(f"⚠️ {file_path}: 분석할 텍스트가 부족합니다.")
        return

    # 정제된 전체 텍스트를 형태소 단위로 토큰화합니다.
    tokens = kiwi.tokenize(clean_text)
    total_tokens = len(tokens)

    # --------------------------------------------------------------
    # 지표 계산
    # --------------------------------------------------------------
    # 1. 활동성 (VV, VA)
    # 동사(VV)와 형용사(VA) 태그를 활동성 지표로 삼습니다.
    dynamic_tags = ('VV', 'VA')
    # 전체 토큰 중 동사/형용사에 해당하는 토큰의 개수를 셉니다.
    dynamic_count = len([t for t in tokens if t.tag.startswith(dynamic_tags)])
    # 전체 토큰 대비 활동성 어휘의 비율(%)을 구합니다.
    activity_index = (dynamic_count / total_tokens) * 100 if total_tokens > 0 else 0

    # 2. 어휘 다양성 (MSTTR 500)
    # 500단어씩 끊어서 중복 없는 어휘의 비율을 계산하는 방식(MSTTR)을 사용합니다.
    window_size = 500
    if total_tokens < window_size:
        # 단어 수가 500개 미만이면 전체 텍스트를 기준으로 계산합니다.
        msttr_score = (len(set(t.form for t in tokens)) / total_tokens) * 100
    else:
        # 500개 이상이면 500개 단위의 묶음별로 다양성을 구하기 위해 빈 리스트를 만듭니다.
        ttr_list = []
        # 전체 길이를 500으로 나눈 몫만큼 반복합니다.
        for i in range(total_tokens // window_size):
            # 500개씩 토큰을 잘라냅니다.
            seg = tokens[i*window_size : (i+1)*window_size]
            # 해당 구간 내 중복 없는 단어의 비율을 구해 리스트에 추가합니다.
            ttr_list.append(len(set(t.form for t in seg)) / window_size)
        # 모든 구간 다양성의 평균(%)을 구합니다.
        msttr_score = np.mean(ttr_list) * 100

    # 3. 참여도 (NP, EC, EF)
    # 대명사, 청유형 어미, 의문형 종결어미 등 독자의 참여를 유도하는 특정 단어들을 정의합니다.
    target_np = {"우리", "너", "그대", "자네", "당신", "너희", "여러분"}
    target_ec = {"ㅂ시다", "ㄹ까요"}
    target_ef = {"ㄹ래요", "ㄹ까", "ㄹ까요", "나요", "ㄴ가", "ㅂ니까", "는가", "을까", "을까요", "ㄴ가요"}
    
    # 조건에 맞는 참여 유도 토큰이 나올 때마다 1씩 더해 총합을 구합니다.
    eng_count = sum(1 for t in tokens if (t.tag == 'NP' and t.form in target_np) or 
                                         (t.tag == 'EC' and t.form in target_ec) or 
                                         (t.tag == 'EF' and t.form in target_ef))
    # 전체 단어 대비 참여 유도 어휘의 비율(%)을 계산합니다.
    engagement_index = (eng_count / total_tokens) * 100 if total_tokens > 0 else 0

    # 4. 감성 지표
    # 앞서 만든 함수를 이용해 분리된 각 문장들의 감성 점수를 리스트로 만듭니다.
    scores = [get_sentiment_score(s) for s in sentences]
    # 전체 문장 감성 점수의 평균을 구합니다.
    mean_sentiment = np.mean(scores)
    # 전체 문장 감성 점수의 표준편차(감성 기복)를 구합니다.
    std_sentiment = np.std(scores)

    # --------------------------------------------------------------
    # 자동 피드백 리포트 출력 (카테고리별 맞춤 조건문)
    # --------------------------------------------------------------
    # 분석 결과를 터미널에 보기 좋게 출력합니다.
    print("\n" + "="*70)
    print(f"📝 자동 분석 리포트: {os.path.basename(file_path)} (카테고리: {category})")
    print("="*70)

    # 카테고리가 'P'인 경우의 평가 기준과 피드백을 적용합니다.
    if category == 'P':
        print("📌 [Pilgrimage (P) 평가 기준: 차분하고 긍정적인 통일된 어휘]")
        
        # 1. 활동성 평가 (Target: 7.35, Bad: 9.55 -> 기준점 8.45)
        # 산출된 활동성 점수가 기준점 이하면 충족, 아니면 주의 상태로 분류합니다.
        act_status = "✅ 충족" if activity_index <= 8.45 else "⚠️ 주의"
        print(f"- [활동성] {act_status}: 현재 {activity_index:.2f}%")
        # 주의 상태일 경우, 구체적인 개선 가이드를 출력합니다.
        if activity_index > 8.45:
            print("  💡 피드백: 동사/형용사 비율이 너무 높습니다(Bad 평균 9.55%). 묘사를 줄이고 수치를 7.35% 수준으로 낮춰보세요.")

        # 2. 다양성 평가 (Target: 43.16, Bad: 45.10 -> 기준점 44.13)
        div_status = "✅ 충족" if msttr_score <= 44.13 else "⚠️ 주의"
        print(f"\n- [어휘 다양성] {div_status}: 현재 {msttr_score:.2f}%")
        if msttr_score > 44.13:
            print("  💡 피드백: 불필요하게 다양한 어휘가 사용되었습니다. 핵심 키워드 위주로 용어를 통일하여 43% 초반으로 맞추세요.")

        # 3. 평균 감성 평가 (Target: 0.471, Bad: 0.277 -> 기준점 0.374)
        sen_status = "✅ 충족" if mean_sentiment >= 0.374 else "⚠️ 주의"
        print(f"\n- [평균 감성] {sen_status}: 현재 {mean_sentiment:.3f}")
        if mean_sentiment < 0.374:
            print("  💡 피드백: 글의 톤이 너무 건조하거나 부정적입니다. 조금 더 긍정적인 뉘앙스의 문장을 추가하세요(목표: 0.47).")

        # 4. 감성 표준편차 평가 (Target: 0.780, Bad: 0.846 -> 기준점 0.813)
        std_status = "✅ 충족" if std_sentiment <= 0.813 else "⚠️ 주의"
        print(f"\n- [감성 기복(표준편차)] {std_status}: 현재 {std_sentiment:.3f}")
        if std_sentiment > 0.813:
            print("  💡 피드백: 문장 간 긍정/부정 감정의 기복이 심합니다. 일관된 톤을 유지하세요(목표: 0.78 이하).")

    # 카테고리가 'T'인 경우의 평가 기준과 피드백을 적용합니다.
    elif category == 'T':
        print("📌 [The Trail (T) 평가 기준: 담백한 소통과 정제된 긍정적 메시지]")

        # 1. 다양성 평가 (Target: 43.56, Bad: 46.85 -> 기준점 45.20)
        div_status = "✅ 충족" if msttr_score <= 45.20 else "⚠️ 주의"
        print(f"- [어휘 다양성] {div_status}: 현재 {msttr_score:.2f}%")
        if msttr_score > 45.20:
            print("  💡 피드백: 어휘가 너무 산만합니다(Bad 평균 46.8%). 불필요한 수식어를 덜어내고 43.5% 수준으로 정제하세요.")

        # 2. 참여도 평가 (Target: 0.872, Bad: 1.212 -> 기준점 1.042)
        eng_status = "✅ 충족" if engagement_index <= 1.042 else "⚠️ 주의"
        print(f"\n- [참여도] {eng_status}: 현재 {engagement_index:.3f}%")
        if engagement_index > 1.042:
            print("  💡 피드백: 대명사(당신, 여러분)나 의문/청유문이 너무 많아 오히려 역효과가 납니다. 0.87% 이하로 담백하게 줄이세요.")

        # 3. 평균 감성 평가 (Target: 0.513, Bad: 0.372 -> 기준점 0.442)
        sen_status = "✅ 충족" if mean_sentiment >= 0.442 else "⚠️ 주의"
        print(f"\n- [평균 감성] {sen_status}: 현재 {mean_sentiment:.3f}")
        if mean_sentiment < 0.442:
            print("  💡 피드백: 메시지가 긍정적으로 전달되지 않고 있습니다. 긍정적인 단어와 결론을 보강하세요(목표: 0.51).")

    # P나 T가 아닌 잘못된 카테고리가 입력되었을 때의 예외 처리입니다.
    else:
        print("알 수 없는 카테고리입니다. 'P' 또는 'T'를 입력해주세요.")

    print("="*70 + "\n")

# ------------------------------------------------------------------
# 실행부 (신규 데이터 P17, T21 평가)
# ------------------------------------------------------------------
# 이 스크립트가 메인으로 실행될 때만 아래 테스트 코드가 작동합니다.
if __name__ == "__main__":
    # 새로 분석할 파일 리스트와 해당 카테고리 매핑 정보를 튜플로 정의합니다.
    new_targets = [
        ("P17.txt", "P"),
        ("T21.txt", "T")
    ]
    
    # 정의된 대상 리스트를 순회하며 분석을 진행합니다.
    for file_name, category in new_targets:
        # 파일이 실제로 존재하지 않을 경우를 대비한 안전 장치 (임시 텍스트 생성)
        # 실제 환경에서는 이 블록을 지우시거나 무시하셔도 됩니다.
        if not os.path.exists(file_name):
            # 파일이 없으면 쓰기 모드('w')로 임시 파일을 생성하여 가짜 텍스트를 채워 넣습니다.
            with open(file_name, "w", encoding="utf-8") as temp_f:
                temp_f.write(f"이것은 {file_name}을 위한 새로운 뉴스레터 테스트 텍스트입니다. " * 15)
                
        # 준비된 파일명과 카테고리를 넣어 리포트 생성 함수를 최종적으로 호출합니다.
        evaluate_new_newsletter(file_name, category=category)
````
